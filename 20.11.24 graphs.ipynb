```python
import pandas as pd
df = pd.read_excel('Dataset1.xlsx')
df
```

### 2. plot a Bar graph to visualize total sales by regiopn. which province had the highest sales made

```python
import matplotlib.pyplot as plt
# Group by 'region' and sum the 'sales'
grouped_data = df.groupby('Region')['Sales'].sum()

x = grouped_data.index.tolist()
y = grouped_data.values.tolist()

plt.bar(x,y, label = 'Bar', width =.5, color = 'black')
plt.title('Highest sales made in provinces')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.legend(loc ='upper right')
for i in range(len(x)):
    plt.text(i, y[i], y[i], ha = 'center')
plt.show()
```

### Use pie chart to show each salesperson's contribution to the total sales. who contributed the most?

```python
import matplotlib.pyplot as plt
salesp = df.groupby('Last Name')['Sales'].sum().reset_index()
plt.figure(figsize=(10,6))
datalabels = salesp['Last Name'].tolist()
values = salesp['Sales'].tolist()
y = min(values)
min_index = values.index(y)
ex_value = ([0 if i!=min_index else 0.2 for i in range(len(values))])
plt.pie(values, labels= datalabels, autopct= '%1.1f%%', explode= ex_value)
plt.title('Salespersons contribution')
plt.show()
```
