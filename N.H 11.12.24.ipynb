### 5. read toothpaste sales data of each month and visualize it using a scatter plot.

### 6. Get the total profit for each month and show a line plot with the following style properties:
- line style dotted and line line color should be red
- show legend at the lower right location
- X label name = Month Number
- Y label name = Units sold
- Circle marker for each month
- line marker color as red
- line width should be 3

```python
import matplotlib.pyplot as plt
import pandas as pd
df = pd.read_csv('Sales.csv')
x = df['month_number'].tolist()
y = df['total_units'].tolist()
plt.plot(x,y, 'blue', label = 'profit/month', linewidth = 3, marker ='o', color = 'red' )
plt.title('Company Profit')
plt.xlabel('Month_number')
plt.ylabel('unit_sold')
plt.grid(True, color = 'pink')
plt.gca().set_facecolor('grey')
plt.legend(loc ='lower right')
plt.show()
```
